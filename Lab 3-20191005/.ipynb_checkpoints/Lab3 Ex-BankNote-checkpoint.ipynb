```python
from sklearn.model_selection import train_test_split
X= df.iloc[:,0:-1].values
y=df.iloc[:,-1].values
x_train,x_test, y_train, y_test=train_test_split(X,y,test_size=0.4, random_state=0)
print(x_train)
print(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix,precision_recall_fscore_support

from sklearn.model_selection import cross_validate
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit_transform(x_train,y_train)
sc.fit(x_test,y_test)
```

```python
dtree = DecisionTreeClassifier().fit(x_train, y_train)
predict = dtree.predict(x_test)
print("The prediction accuracy is: {0:2.2f}{1:s}".format(dtree.score(x_test,y_test)*100,"%"))
```

```python
dtree = DecisionTreeClassifier(criterion = 'entropy').fit(x_train, y_train)
predict = dtree.predict(x_test)
print("The prediction accuracy is: {0:2.2f}{1:s}".format(dtree.score(x_test,y_test)*100,"%"))
```

```python
cm = confusion_matrix(y_test, predict)
cm_df = pd.DataFrame(cm, index =['0','1'],
columns =['0','1'])

plt.figure(figsize=(7,4))
sns.heatmap(cm_df, annot=True)
plt.title("Decision Tree Accuracy:" + str(dtree.score(x_test,y_test)*100))
plt.ylabel('True label')
plt.xlabel('Predicted label')
```

```python
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,
filled=True, rounded=True,
special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```python
from sklearn.model_selection import cross_val_score 
depth=[]
Irange=list(range(1,15))
for i in Irange:
    clf = DecisionTreeClassifier(max_depth=i)
    scores = cross_val_score(clf, x_train, y_train, cv=5, scoring='accuracy')
    depth.append((scores.mean()))
    
optimal_i=Irange[depth.index(max(depth))]    
clf=DecisionTreeClassifier(max_depth=optimal_i).fit(x_train,y_train)
predict= clf.predict(x_test)
print("The optimal depth is %d" % optimal_i)
#print("The optimal scores is %.2f" %max(depth))
print("The prediction accuracy is:{0:2.5f}".format(clf.score(x_test,y_test)*100,"%"))

plt.figure(2,figsize=(15,5))

plt.plot(Irange, depth,ls='dashed')
plt.xlabel("Depth of Decision Tree")
plt.ylabel('Accuracy Score')
plt.title('Accuracy score for Depth of Decision Tres')


```
