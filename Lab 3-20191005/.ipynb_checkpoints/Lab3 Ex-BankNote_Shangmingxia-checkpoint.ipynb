## Step 1. Data preprocessing

### Step1.1 Get correlation

```python
# correlations using heat map plot
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True,cmap='winter')
plt.show()
```

```python
features = df.iloc[:,:-1]
labels = df.iloc[:, -1]
```

### Step1.2 Scale data && get testdata and traindata

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.4,random_state=0)
scaller = StandardScaler()
x_train = scaller.fit_transform(x_train)
x_test = scaller.transform(x_test)
```

## Step 2. Prediction

Use default parameter configurations.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix,precision_recall_fscore_support
dtree = DecisionTreeClassifier().fit(x_train, y_train) 
predict = dtree.predict(x_test)
print("The prediction accuracy is: {0:2.5f}{1:s}".format(dtree.score(x_test,y_test)*100,"%")) 
```

```python
dtree = DecisionTreeClassifier(criterion = 'entropy').fit(x_train, y_train) 
print("The prediction accuracy is: {0:2.5f}{1:s}".format(dtree.score(x_test,y_test)*100,"%")) 
```

Q2:Try using 1 other parameter configurations and compare them to their default configuration.
How much they differ?

about 0.6%

```python
from sklearn.model_selection import cross_val_score 
score = cross_val_score(dtree, x_train, y_train, cv=15) 
print("The prediction accuracy is: {0:2.5f}{1:s}".format(score.mean()*100,"%")) 
```

## step3 cross validation

```python
scores = []
lrange=list(range(1,15))
for i in lrange:
    dtree = DecisionTreeClassifier(max_depth=i).fit(x_train, y_train) 
    score = cross_val_score(dtree, x_train, y_train, cv=10, scoring='accuracy') 
    scores.append(score.mean())
optimal_i = lrange[scores.index(max(scores))]
dtree1 = DecisionTreeClassifier(max_depth=optimal_i).fit(x_train, y_train) 
print(optimal_i)
print("The prediction accuracy is: {0:2.5f}{1:s}".format(dtree1.score(x_test,y_test)*100,"%")) 
```
