```python
from datetime import datetime
import pandas as pd
df = pd.read_csv('Emp.csv')
# replace '-' symbol with '/'
df['hiredate'] = df['hiredate'].str.replace('-','/')
df['hiredate']=pd.to_datetime(df['hiredate'],dayfirst=True)
df['hiredate']=df['hiredate'].dt.strftime('%d/%m/%Y')
df['hiredate']
```
