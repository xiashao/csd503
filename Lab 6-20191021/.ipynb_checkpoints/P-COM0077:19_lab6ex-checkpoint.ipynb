```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from scipy import stats
# use seaborn plotting defaults
import seaborn as sns; 
import warnings
warnings.filterwarnings("ignore")
sns.set()
```

```python
data=pd.read_csv('Iris.csv')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
scaller = StandardScaler()
labelencoder = LabelEncoder()
data['Species']= labelencoder.fit_transform(data['Species'])
```

```python
x = data.iloc[:, 0:5]
y=data.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
scaller = StandardScaler()
x_train = scaller.fit_transform(x_train)
x_test = scaller.transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
rate=0.01
rates=[]
scores=[]
for i in range(0,100):
    model = MLPClassifier(learning_rate_init=rate)
    model.fit(x_train, y_train)
    rates.append(model.learning_rate_init)
    rate=0.01+rate
    score=model.score(x_test,y_test)
    scores.append(score)
    #print(model.learning_rate_init)
    #print("The prediction accuracy is: {0:2.5f}{1:s}".format(model.score(x_test,y_test)*100,"%")) 
plt.figure(2,figsize=(15,5))
plt.plot(rates, scores,ls='dashed')
plt.xlabel('Value of learning_rate')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores for Values of learning_rate')
plt.show()
```

```python
iteration=100
max_iters=[]
scores=[]
for i in range(0,10):
    model = MLPClassifier(max_iter=iteration)
    model.fit(x_train, y_train)
    max_iters.append(model.max_iter)
    iteration=100+iteration
    score=model.score(x_test,y_test)
    scores.append(score)
plt.figure(2,figsize=(15,5))
plt.plot(max_iters, scores,ls='dashed')
plt.xlabel('Value of iteration number')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores for Values of teration number')
plt.show()
```
