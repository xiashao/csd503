```python
df.drop(['Specimen','Lobedness','Average_contrast','Smoothness','Third_moment','Maximal_indentationDepth',
        'Isoperimetic_factor','Stochastic_convexity','Eccentricity'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
X=df.iloc[:,1:]
labelencoder=LabelEncoder()
Y=labelencoder.fit_transform(y)
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.4, random_state=30)


gnb =GaussianNB()
gnb.fit(x_train,y_train)
predict = gnb.predict(x_test)
print("Number of mislabeled points out of a total {} points : {}, performance {:05.2f}%"
.format(x_test.shape[0], (y_test != predict).sum(),
gnb.score(x_test,y_test)*100 ))
```

```python
cm=confusion_matrix(y_test,predict)
cm_df = pd.DataFrame(cm)

plt.figure(figsize=(20,5))
sns.heatmap(cm_df, annot=True, fmt='g')
plt.title("Naive Bayes Accuracy:" + str(gnb.score(x_test,y_test)*100))
plt.ylabel('True label')
plt.xlabel('Predicted label')
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

scores=[]
Irange=list(range(1,30))
for k in Irange:
    knn=KNeighborsClassifier(n_neighbors=k,weights='distance',algorithm='auto')
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    scores.append(metrics.accuracy_score(y_test,y_pred))
    
plt.figure(2,figsize=(15,5))

plt.plot(Irange, scores,ls='dashed')
plt.xlabel("Value of K for KNN")
plt.ylabel('Accuracy Score')
plt.title('Accuracy score for values of K of K-Nearest-Neighbors')

optimal_k=Irange[scores.index(max(scores))]    
print("The optimal number of neighbors is %d" % optimal_k)
print("The optimal score is %.2f" %max(scores))
    
knn_model=KNeighborsClassifier(optimal_k)
knn_model.fit(x_train,y_train)
print("Accuracy for predict the Test Data is %.2f" % knn_model.score(x_test,y_test))  
```

```python
from sklearn.model_selection import cross_val_score
scores=[]
Irange=list(range(1,20))
for k in Irange:
    knn=KNeighborsClassifier(n_neighbors=k,weights='distance',algorithm='auto')
    score=cross_val_score(knn,x_train,y_train,cv=3,scoring='accuracy')
    scores.append(score.mean())
    
optimal_k=Irange[scores.index(max(scores))]
print("The optimal number of neighbors is %d" % optimal_k)
print("The optimal score is %.2f"%max(scores))

plt.figure(2,figsize=(15,5))
plt.plot(Irange,scores,ls='dashed')
plt.xlabel('Value of K of KNN')
plt.ylabel('Accuracy Score')
plt.title('Accuracy score for values of K of K-Nearest-Neighbors')

knn_model=KNeighborsClassifier(optimal_k)
knn_model.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix,precision_recall_fscore_support
from sklearn.model_selection import cross_validate

dtree = DecisionTreeClassifier(criterion = 'entropy').fit(x_train, y_train)
predict = dtree.predict(x_test)
print("The prediction accuracy is: {0:2.2f}{1:s}".format(dtree.score(x_test,y_test)*100,"%"))
```

```python
from sklearn.model_selection import cross_val_score 
depth=[]
Irange=list(range(1,15))
for i in Irange:
    clf = DecisionTreeClassifier(max_depth=i)
    scores = cross_val_score(clf, x_train, y_train, cv=3, scoring='accuracy')
    depth.append((scores.mean()))
    
optimal_i=Irange[depth.index(max(depth))]    
clf=DecisionTreeClassifier(max_depth=optimal_i).fit(x_train,y_train)
predict= clf.predict(x_test)
print("The optimal depth is %d" % optimal_i)
#print("The optimal scores is %.2f" %max(depth))
print("The prediction accuracy is:{0:2.5f}".format(clf.score(x_test,y_test)*100,"%"))

plt.figure(2,figsize=(15,5))

plt.plot(Irange, depth,ls='dashed')
plt.xlabel("Depth of Decision Tree")
plt.ylabel('Accuracy Score')
plt.title('Accuracy score for Depth of Decision Tres')
```
