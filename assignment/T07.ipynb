## Step4  get testdata and traindata

```python
from sklearn.model_selection import train_test_split
y = df['class']
x = df.drop(columns=['class'],axis=1)
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size = 0.4)
```

## Step5 Predition and get performance

```python
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, auc, roc_curve
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz
```

```python
import numpy as np
dtree = DecisionTreeClassifier().fit(x_train, y_train) 
print("The prediction accuracy is: {0:2.5f}{1:s}".format(dtree.score(x_test,y_test)*100,"%")) 
```

```python
y_pred=dtree.predict(x_test)
print("Decision Tree Classifier report \n", classification_report(y_test, y_pred))
```

```python
cm=confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm)
sns.heatmap(cm_df, annot=True, fmt='g')
plt.title('Decision Tree Classifier confusion matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train,y_train)
# Test the classifier
predict = gnb.predict(x_test)
print("The prediction accuracy is: {0:2.5f}{1:s}".format(gnb.score(x_test,y_test)*100,"%")) 
```

```python
print("Bayes Classifier report \n", classification_report(y_test, predict))
```

```python
cm=confusion_matrix(y_test, predict)
cm_df = pd.DataFrame(cm)
sns.heatmap(cm_df, annot=True, fmt='g')
plt.title('Bayes Classifier confusion matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
```

## Step6 Comparision

```python
from sklearn.metrics import accuracy_score
dtree = DecisionTreeClassifier()
gnb = GaussianNB()
knn = KNeighborsClassifier()
svm=SVC()
classifiers = [dtree,gnb,knn,svm]
classifier_names = ['Decision Tree','Bayes','KNN','SVM']
precisions=[]
recalls = []
for classifier,classifier_name in zip(classifiers,classifier_names):
    classifier.fit(x_train,y_train)
    predicted_labels = classifier.predict(x_test)
    precision= accuracy_score(y_test, predicted_labels)
    recall=metrics.recall_score(y_test, predicted_labels, average='micro')
    recalls.append(recall)
    precisions.append(precision)
labels = ['recall','precision']
pfs = [recalls,precisions]
ii=[1,2]
plt.figure(2,figsize=(15,5))
for label1,pf,i in zip(labels,pfs,ii):
    plt.subplot(1,2,i)
    plt.bar(classifier_names, pf, label=label1)
    plt.xlabel("classifier_names")
    plt.ylabel(label1)
    plt.legend(loc=4)
```

## Step7 Cross Validation

```python
from sklearn.model_selection import cross_val_score 
scores=[]
for classifier,classifier_name in zip(classifiers,classifier_names):
    score = cross_val_score(classifier, x_train, y_train,cv=10) 
    scores.append(score.mean())
plt.bar(classifier_names, scores)
plt.xlabel("classifier_names")
plt.ylabel("accuracy")
plt.legend(loc=4)
plt.show()
```

```python
from sklearn.model_selection import cross_val_score 
depth=[]
Irange=list(range(1,15))
for i in Irange:
    clf = DecisionTreeClassifier(max_depth=i)
    scores = cross_val_score(clf, x_train, y_train, cv=5, scoring='accuracy')
    depth.append((scores.mean()))   
optimal_i=Irange[depth.index(max(depth))]    
clf=DecisionTreeClassifier(max_depth=optimal_i).fit(x_train,y_train)
predict= clf.predict(x_test)
print("The optimal depth is %d" % optimal_i)
print("The prediction accuracy is:{0:2.5f}".format(clf.score(x_test,y_test)*100,"%"))
plt.figure(2,figsize=(15,5))
plt.plot(Irange, depth,ls='dashed')
plt.xlabel("Depth of Decision Tree")
plt.ylabel('Accuracy Score')
plt.title('Accuracy score for Depth of Decision Tres')
```
